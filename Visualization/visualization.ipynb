# Cartesian Product and Expansion
## 1D, 2D, and 3D

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-2, 2, 200)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot(x, np.zeros_like(x), 'k-', linewidth=1)
plt.scatter(x, np.zeros_like(x), c=x, cmap='rainbow', s=5)
plt.xlabel('x')
plt.title('1D Space')
plt.grid(True)


x, y = np.meshgrid(x, x)

plt.subplot(1, 2, 2)
plt.scatter(x, y, c=x, cmap='rainbow', s=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Space (Cartesian Product)')
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 12))
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
z = np.linspace(-2, 2, 50)
x, y, z = np.meshgrid(x, y, z)

ax = plt.subplot(1, 3, 3, projection='3d')
ax.scatter(x, y, z, c=x, cmap='rainbow', s=5, alpha=0.1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Space (Cartesian Product)')
ax.view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()
```

```python
#f(x, y) = x + y
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = X + Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Graph of f(x, y) = x + y')
plt.show()
```

## Function VS Relation

```python
import matplotlib.pyplot as plt
import numpy as np

# Values
r = 1  # radius of the circle
x = np.linspace(-r, r, 400)
y1 = np.sqrt(r**2 - x**2)
y2 = -np.sqrt(r**2 - x**2)

# Plotting the circle (relation)
plt.figure(figsize=(8, 8))
plt.plot(x, y1, 'blue', label='$y = \sqrt{r^2 - x^2}$')
plt.plot(x, y2, 'blue')  # The other half of the circle

# Plotting the function y = x^2
x_func = np.linspace(-r, r, 400)
y_func = x_func**2
plt.plot(x_func, y_func, 'red', label='$y = x^2$')

# Labels and title
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Set equal scaling
plt.axis('equal')

# Display the plot
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()


```
